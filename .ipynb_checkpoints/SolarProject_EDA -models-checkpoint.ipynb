```python
#wind direction is in degrees.in reality,0 degree and 360 degree is same and distance between 2 degree and 358 degree is small.but a model doesnt 
#know this it thinks distance between 2 degree and 358 degree is 356.this will decrease the accuracy of model.so we convert degrees to points using 
#sine and cosine transformation
```

```python
data1["wind-direction"] =data1["wind-direction"]*10 #in dataset 360 is given as 36,270 is given as 27.so we multiply them with 10.
```

```python
data1["wind-dir-rad"] = np.deg2rad(data1["wind-direction"])
data1["wind-dir-sin"] = np.sin(data1["wind-dir-rad"])
data1["wind-dir-cos"] = np.cos(data1["wind-dir-rad"])
```

```python
data1.drop(columns=["wind-direction","wind-dir-rad"],inplace=True)
```

```python
X=data1.drop('power-generated',axis=1)
y=data1['power-generated']  #power generated is target variable
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40) 
#this is used for tree based models cuz for tree based models, transfomers and scaling is not required.and encoding(labelencoder) is required if text 
#data is there but here data is in numeric.
```

```python
data1["sky-cover"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe = OneHotEncoder(drop='first',sparse_output=False)
encoded = oe.fit_transform(data1[['sky-cover']])
encoded_col = pd.DataFrame(encoded, columns=oe.get_feature_names_out(['sky-cover']))
encoded_data = pd.concat([data1.drop('sky-cover', axis=1),encoded_col],axis=1)
```

```python
encoded_data
```

```python
cols=["distance-to-solar-noon","temperature","wind-speed","sky-cover","visibility","humidity","average-wind-speed-(period)","average-pressure-(period)","power-generated"]
for col in cols:
    sns.histplot(data[col],kde=True,bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
data1.skew()
#from above histogram with kde and here skewness values, we can see visibility and humidity are left skewed so we use powertransformer to make it 
#less skewed and power genarated is right skewed,so we use log tranformation to make it less skewed.
#Note:we can use transformer on target variable.how it helps,you will see at the end.
```

```python
X_encoded = encoded_data.drop("power-generated",axis=1)
y_encoded = encoded_data["power-generated"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_encoded,y_encoded,test_size=0.2,random_state=40)
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer()
X_train[["visibility","humidity"]] = pt.fit_transform(X_train[["visibility","humidity"]])
X_test[["visibility","humidity"]] = pt.transform(X_test[["visibility","humidity"]])
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
cols=["distance-to-solar-noon","temperature","wind-speed","average-wind-speed-(period)","average-pressure-(period)"]
X_train[cols] = scaler.fit_transform(X_train[cols])
X_test[cols] = scaler.transform(X_test[cols])
```

```python
X_train
```

```python
X_train.describe()
```

```python
#here we can observe that we have used powertransformer on visibility and humidity columns which means these columns range wil be -2 to +2.
#and for wind direction we have used sin and cos tranformation so the range will be -1 to +1
#and for remaining columns we have used standardscalar which means there range will be -3 to +3.
#if we look all the columns at a time.we may find ranges are different but all columns mean is around 0 and standard deviation around 1 which makes
#our data is absolutely perfect for model fitting.
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
r2_score(y_test, y_pred)
```

```python
y_train_log = np.log1p(y_train)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train_log)
```

```python
y_pred_log=lr.predict(X_test)
y_pred = np.expm1(y_pred_log)
r2_score(y_test,y_pred)
```

```python
y_pred_log=lr.predict(X_test)
r2_score(np.log1p(y_test), y_pred_log)
```

```python
#when r2 score is calculated between log values of y test and y pred,the score is 81 but when we inverse that using exponential ,score is -89 it is
#cuz exp means (10^) so the value will become like 36000 instead of 36058.this makes r2 score very bad.so we use only log values to check r2 score
#and while predicting new values,we will convert predicted value using np.exmp1.
```
